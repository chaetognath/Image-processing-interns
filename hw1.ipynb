4-1. Generate the bit planes of lena image acquired in 3.1) and display all the bit planes in a 
single figure.  

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io

w=io.imread('images\lena.bmp')
bps= [(w>>i)%2 for i in range(8)]
for i in range(8):
    plt.subplot(3,3,1+i)
    plt.imshow(bps[i], cmap='gray')
    plt.axis('off')
```
