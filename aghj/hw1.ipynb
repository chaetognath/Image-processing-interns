3-2. Reduce the image size(rows and columns respectively) by 1/2, 1/4, 1/8. (Use the parameter 
order=0 for skimage.transform.rescale())

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
import skimage.transform as hj

w=io.imread('images\lena.bmp')

plt.subplot(2, 2, 1)
plt.title("1/2")
z = hj.rescale(w, 1/2)
plt.imshow(z, cmap='gray')
plt.show()

plt.subplot(2, 2, 2)
plt.title("1/4")
z = hj.rescale(w, 1/4)
plt.imshow(w, cmap='gray')
plt.show()

plt.subplot(2, 2, 3)
plt.title("1/8")
z = hj.rescale(w, 1/8)
plt.imshow(w, cmap='gray')
plt.show()
```

4-1. Generate the bit planes of lena image acquired in 3.1) and display all the bit planes in a 
single figure.  

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io

w=io.imread('images\lena.bmp')
bps= [(w>>i)%2 for i in range(8)]
for i in range(8):
    plt.subplot(3,3,1+i)
    plt.imshow(bps[i], cmap='gray')
    plt.axis('off')
```

4-2. Combine only two MSB planes together into one image. 

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io

w=io.imread('images\lena.bmp')
bps= [(w>>i)%2 for i in range(8)]
#range(4)로 해도 No matter~~
z=sum((bps[i]*2**i for i in range(6,8)))
(w==z).all()
plt.figure()
plt.imshow(z, cmap='gray')
plt.axis('off')
plt.show()
```
